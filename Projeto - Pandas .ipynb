Calcule o percentual de medalhas de ouro, prata e bronze que o Brasil ganhou em cada olimpíada.

````python
import pandas as pd
#df = pd.read_csv(r'C:\Users\Samsung\Desktop\Vida\Carreira\cientista de dados\lets code\exs\Listas\athlete_events.csv')
df = pd.read_csv(r'athlete_events.csv')
df_brasil = df.loc[df['Team'] == 'Brazil'].copy()

df_brasil.dropna(axis = 0 , subset = ['Medal'], inplace = True  )

#df_brasil.loc[df_brasil.Year == 1920].drop_duplicates(subset = ['Event'], keep = 'first')
df_brasil.Medal = pd.Categorical(values=df_brasil.Medal, categories=['Bronze', 'Silver', 'Gold'], ordered=True)
df_brasil.drop_duplicates(subset=['Event','Games'], keep = 'first', inplace = True)
brasil_por_ano = df_brasil.groupby(['Games', 'Medal'])['Name'].count()
brasil_por_ano = brasil_por_ano.unstack(level = 'Medal')


for i in brasil_por_ano.index: 
    
    if brasil_por_ano.loc[i].sum(axis = 0) > 0:
        
        sum_year = brasil_por_ano.loc[i].sum(axis = 0)
        
        brasil_por_ano.loc[i]= brasil_por_ano.loc[i].astype('float') /sum_year
        
        


brasil_por_ano*=100
brasil_por_ano
````
